**TIME COMPARISON BETWEEN PYTHON ARRAY AND NUMPY ARRAYS**

````python
current_time = time.time()
current_time
````

````python
import time
size = 10000
def pure_python_version():
  current_time = time.time()
  X = range(size)
  Y = range(size)
  Z = [X[i] + Y[i] for i in range(len(X)) ]
  return time.time() - current_time
````

````python
import numpy as np
def numpy_version():
  current_time = time.time()
  X = np.arange(size)
  Y = np.arange(size)
  Z = X + Y
  return time.time() - current_time
````

# **Ratio of time taken by numpy array to python array**

````python
t1 = pure_python_version()
t2 = numpy_version()
print(t1, t2)
print(t1/t2)
````

numpy `arange()` function creates an array of evenly spaced values within a given interval. It is similar to Python's built-in `range()` function but returns a NumPy array instead of a list.

````python
arr = np.arange(1,5) #np.arange(start, end, step)
arr
````

Adding a step argument in `arange()` allowed the generation of array with numbers from start to end with steps of `size = 2`

````python
arr = np.arange(0,100,2)
arr
````

````python
arr = np.arange(100,0,-2)  ##changing the start point and end point allows you to take reverse steps
arr
````

Another interesting inbuilt function of numpy is `linspace()` returns an array of evenly spaced numbers over a specified range. Unlike the range() function in Python that generates numbers with a specific step size. linspace() allows you to specify the total number of points you want in the array, and NumPy will calculate the spacing between the numbers automatically.

````python
arr = np.linspace(0,1,5)
arr
````

DataFrames can be indexed by column name (label) or row name (index) or by the serial number of a row. The loc method is used for indexing by name, while iloc() is used for indexing by number.

In the first case below, we say "give us the values of the rows with index from 0 to 5 (inclusive) and columns labeled from Direction to Country (inclusive)".

````python
data.loc[0:5, "trip_distance":"total_amount"] #column based indexing
````

In the second case, we say "give us the values of the first five rows in the first three columns" (as in a typical Python slice: the maximal value is not included).

````python
data = data.iloc[0:5,0:3] #integer-location based indexing
#data
````
